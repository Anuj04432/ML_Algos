```python
from sklearn.datasets import make_classification
X,y = make_classification(n_samples=500,n_features=10,n_informative =2,n_redundant=5,class_sep=0.5,flip_y=0.15 ,n_classes=2,random_state=42)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)  
```

```python
X,y
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
dummy_pred = [0 for i in range(len(y_test))]
```

```python
dummy_pred
```

```python
y_prob = model.predict_proba(X_test)[:,1]
y_prob
```

```python
from sklearn.metrics import roc_curve,roc_auc_score
```

```python
fpr,tpr,thresholds = roc_curve(y_test,y_prob)
dummy_fpr,dummy_tpr,dummy_thresholds = roc_curve(y_test,dummy_pred)
```

```python
# Roc_auc_score
print("Roc_auc_score for model is: ",roc_auc_score(y_test,y_prob))
print("Roc_auc_score for dummy model is: ",roc_auc_score(y_test,dummy_pred))
```

```python
import matplotlib.pyplot as plt
import numpy as np  
import seaborn as sns   
```

```python
# Plotting the roc curve for the model and dummy model
plt.figure(figsize=(6,6))
plt.plot(fpr,tpr,label="Model")
plt.plot(dummy_fpr,dummy_tpr,label="Dummy Model",linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```

```python
# With threshold values and AUC score
auc = roc_auc_score(y_test, y_prob)

plt.figure(figsize=(6,6))
plt.plot(fpr, tpr, label=f"Logistic Regression (AUC = {auc:.3f})")
plt.plot(dummy_fpr, dummy_tpr, '--', label="Dummy Model")
for i in range(0, len(thresholds), 3):   # Every 3rd threshold
    plt.annotate(f'{thresholds[i]:.2f}',
                 (fpr[i], tpr[i]),
                 xytext=(6,6),
                 textcoords='offset points',
                 fontsize=5,
                 bbox=dict(boxstyle="round", fc="white", alpha=0.7))
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve with every 3rd Threshold Value",)
plt.legend(loc="lower right")
plt.grid(True)
plt.show()

```
